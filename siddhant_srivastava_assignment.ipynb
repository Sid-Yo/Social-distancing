```
def generateTimeKeys():

  timings = []
  for i in range(48):
    if(i == 0):
      timings.append('00:00')
    elif(i%2 == 0 and i!= 0):
      hour = i//2
      if(hour <= 9):
        timings.append('0'+str(hour)+':'+'00')
      else:
        timings.append(str(hour)+':'+'00')
    else:
      hour = timings[-1].split(':')[0]
      timings.append(hour+':'+'30')

  return timings
```
